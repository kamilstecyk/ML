```python

# we get sets
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error

size = 300
X = np.random.rand(size)*5-2.5
w4, w3, w2, w1, w0 = 1, 2, 1, -4, 2
y = w4*(X**4) + w3*(X**3) + w2*(X**2) + w1*X + w0 + np.random.randn(size)*8-4 

df = pd.DataFrame({'x': X, 'y': y}) 
df.to_csv('dane_do_regresji.csv',index=None)
df.plot.scatter(x='x',y='y')

dfSorted = df.sort_values(by=['y'])


#dfSorted


```

```python

# we divide sets, we will have training set and test set

from sklearn.model_selection import train_test_split

# df['x'] to bedzie series, natomiast df[['x']] to bedzie pojedynczy df
X_train, X_test, y_train, y_test = train_test_split( dfSorted[['x']], dfSorted[['y']], test_size=0.20, random_state=42)



```

```python


# linear regression and display results

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)

print(lin_reg.intercept_, lin_reg.coef_, "\n")



regression_Predicts = lin_reg.predict(X_test)


trainPredicts =  lin_reg.predict(X_train)

# MSE 
mseTrainLin = mean_squared_error(y_train, trainPredicts)
mseTestLin = mean_squared_error(y_test, regression_Predicts)


#print(regression_Predicts)

plt.scatter(X_test,y_test,color="black")
#plt.scatter(X_test,regression_Predicts,color="blue")
plt.plot(X_test,regression_Predicts,linewidth = 3)

plt.show()






```

```python

# K - nearests neighbours / 3 neighbours

import sklearn.neighbors

knn_3_reg = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)

knn_3_reg.fit(X_train,y_train)

KNN_result = knn_3_reg.predict(X_test)
KNN_train_results = knn_3_reg.predict(X_train)

plt.scatter(X_test,y_test,color="black")
plt.scatter(X_test,KNN_result,color="blue")


# MSE 
mseTrainK3 = mean_squared_error(y_train, KNN_train_results)
mseTestK3 = mean_squared_error(y_test, KNN_result)



```

```python

# K - nearests neighbours / 5 neighbours


knn_5_reg = sklearn.neighbors.KNeighborsRegressor(n_neighbors=5)

knn_5_reg.fit(X_train,y_train)

KNN_result2 = knn_5_reg.predict(X_test)
KNN_result2_train = knn_5_reg.predict(X_train)

plt.scatter(X_test,y_test,color="black")
plt.scatter(X_test,KNN_result2,color="blue")


# MSE 
mseTrainK5 = mean_squared_error(y_train, KNN_result2_train)
mseTestK5 = mean_squared_error(y_test, KNN_result2)

```

```python

# polynomial regression  2 degree 

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression


poly_2_reg = PolynomialFeatures(degree=2, include_bias=False)
poly_feature_2 = poly_2_reg.fit_transform(X_train)



lin_reg2 = LinearRegression()
lin_reg2.fit(poly_feature_2, y_train)


# Visualising the data

#plt.scatter(X_test, y_test, color = 'blue')

# we need to sort data because of plotting later

X_test = X_test.sort_values(by=['x'])
y_test.reset_index()
y_test = y_test.reindex(X_test.index)



plt.scatter(X_test,y_test,color="black")
predictionsResult2 = lin_reg2.predict(poly_2_reg.fit_transform(X_test))
predictionsTrainResult = lin_reg2.predict(poly_2_reg.fit_transform(X_train))
plt.plot(X_test, predictionsResult2)

    

# MSE 
mseTrainPoly2 = mean_squared_error(y_train, predictionsTrainResult)
mseTestPoly2 = mean_squared_error(y_test, predictionsResult2)
```

```python
# polynomial 3 degree

poly_3_reg = PolynomialFeatures(degree=3, include_bias=False)
poly_feature_3 = poly_3_reg.fit_transform(X_train)



lin_reg2 = LinearRegression()
lin_reg2.fit(poly_feature_3, y_train)


# Visualising the data

#plt.scatter(X_test, y_test, color = 'blue')

# we need to sort data because of plotting later


plt.scatter(X_test,y_test)
predictionsResult2 = lin_reg2.predict(poly_3_reg.fit_transform(X_test))
predictionsTrainResult = lin_reg2.predict(poly_3_reg.fit_transform(X_train))
plt.plot(X_test, predictionsResult2)

    
plt.show()

# MSE 
mseTrainPoly3 = mean_squared_error(y_train, predictionsTrainResult)
mseTestPoly3 = mean_squared_error(y_test, predictionsResult2)
```

```python
# polynomial 4 degree

poly_4_reg = PolynomialFeatures(degree=4, include_bias=False)
poly_feature_4 = poly_4_reg.fit_transform(X_train)



lin_reg2 = LinearRegression()
lin_reg2.fit(poly_feature_4, y_train)


# Visualising the data


plt.scatter(X_test,y_test)
predictionsResult2 = lin_reg2.predict(poly_4_reg.fit_transform(X_test))
predictionsTrainResult = lin_reg2.predict(poly_4_reg.fit_transform(X_train))
plt.plot(X_test, predictionsResult2)

    
plt.show()

# MSE 
mseTrainPoly4 = mean_squared_error(y_train, predictionsTrainResult)
mseTestPoly4 = mean_squared_error(y_test, predictionsResult2)
```

```python
# polynomial 5 degree

poly_5_reg = PolynomialFeatures(degree=5, include_bias=False)
poly_feature_5 = poly_5_reg.fit_transform(X_train)



lin_reg2 = LinearRegression()
lin_reg2.fit(poly_feature_5, y_train)


# Visualising the data

#plt.scatter(X_test, y_test, color = 'blue')

# we need to sort data because of plotting later


plt.scatter(X_test,y_test)
predictionsResult2 = lin_reg2.predict(poly_5_reg.fit_transform(X_test))
predictionsTrainResult = lin_reg2.predict(poly_5_reg.fit_transform(X_train))


plt.plot(X_test, predictionsResult2)

    
plt.show()

# MSE 
mseTrainPoly5 = mean_squared_error(y_train, predictionsTrainResult)
mseTestPoly5 = mean_squared_error(y_test, predictionsResult2)
```
