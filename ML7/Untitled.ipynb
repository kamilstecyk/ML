```python
# we load the data

from sklearn.datasets import fetch_openml 
import numpy as np

mnist = fetch_openml('mnist_784', version=1, as_frame=False) 
mnist.target = mnist.target.astype(np.uint8)
X = mnist["data"]
y = mnist["target"]

```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score


listToPickle = []

for i in range(8,13):
    
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    
    score = silhouette_score(X, kmeans.labels_)
    
    listToPickle.append(score)
    print(score)
    print("\n")
    
    
print(listToPickle)
```

```python
# we pickle our list

import pickle

with open('kmeans_sil.pkl', 'wb') as f:
    pickle.dump(listToPickle, f)
    
    
    
```

```python
# we create matrix of errors

# in each row we should have one column which have high result, we are not interersted in diagonal values because it can not be classified in sequence

from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans



kmeans = KMeans(n_clusters=10)
kmeans.fit(X)

yPredicted = kmeans.predict(X)

matrix = confusion_matrix(y,yPredicted)



```

```python
# we pickle our list

#print(matrix)

#import pickle

# this is arbitrary

#with open('matrix.pkl', 'wb') as f:
 #   pickle.dump(matrix, f)
    
    
    
```

```python
# we find max indexes in each row

setToPickle = set()

for row in matrix:
    maxIndexInRow = np.argmax(row)
    setToPickle.add(maxIndexInRow)
    
    
print(setToPickle)

```

```python
# we pickle 

import pickle

with open('kmeans_argmax.pkl', 'wb') as f:
    pickle.dump(setToPickle, f)
    
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

list2ToPickle = []

#dbscan = DBSCAN(eps=0.20, min_samples=5)
#dbscan.fit(X)

for x1 in X[:300]:
    for x2 in X:
        distance = np.linalg.norm(x1-x2)
        list2ToPickle.append(distance)

#list2ToPickle
```

```python
# we need to delete 0 from list , because distance of element from itself is 0 and this is senseless info for us

list2WithoutZeros = [i for i in list2ToPickle if i != 0 ] # comprehension list
list2WithoutZeros.sort()  # we sort ascending

listOfTenMins = list2WithoutZeros[:10] # we use slice

#listOfTenMins
```

```python
# we pickle 

import pickle

with open('dist.pkl', 'wb') as f:
    pickle.dump(listOfTenMins, f)
    

```

```python
# we calc mean for later dbscan alg eps

import numpy as np

s = np.mean(listOfTenMins[:3])  # we take three min values to calc mean

eps1 = s
eps2 = 1.04 * s
eps3 = 1.08 * s

eps = [eps1,eps2,eps3]

```

```python
# we use db scan alg
from sklearn.cluster import DBSCAN

listOfUqLabels = []

for e in eps:
    dbscan=DBSCAN(eps=e)
    dbscan.fit(X)
    labelsSet = set(dbscan.labels_.flat)
    listOfUqLabels.append(len(labelsSet))
    print(labelsSet)
    
print(listOfUqLabels)
```

```python
# we pickle 

import pickle

with open('dbscan_len.pkl', 'wb') as f:
    pickle.dump(listOfUqLabels, f)
    


```
